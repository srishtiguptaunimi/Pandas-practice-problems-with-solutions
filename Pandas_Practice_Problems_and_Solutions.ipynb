```python
#df.loc is used to access any row based on their index
#df.iloc allows to access rows based on

#a subset of both rows and columns is made in one go and just using
#selection brackets [] is not sufficient anymore.
#The loc/iloc operators are required in front of the selection brackets [].
#When using loc/iloc, the part before the comma is the rows you want,
#and the part after the comma is the columns you want to select.
```

#Q12. How to use Conditional statements for filtering records?

```python
data.loc[data['Age'] > 70] #Here original dataset is filtered based on the condition
```
